```python
import pandas as pd
import requests
```

```python
data = []
offset = 0

while True:
    params = {
        "$where" : "created_date >= '2024-01-01T00:00:00.000' AND created_date <= '2025-01-01T00:00:00.000'",
        "$limit" : 2000,
        "$offset" : offset,
        "$order" : 'created_date'
    }
    
    r = requests.get(
        "https://data.cityofnewyork.us/resource/erm2-nwe9.json",
        params = params
    )
    r.raise_for_status()
    
    batch = r.json()
    if len(batch) == 0:
        print("No more Data")
        break
    
    data.extend(batch)
    offset+=2000
    print(f"Fetched {len(data)} rows")
    
    if len(data) >= 20000:
        print("Reached row limit.")
        break
```

```python
df = pd.DataFrame(data)
```
